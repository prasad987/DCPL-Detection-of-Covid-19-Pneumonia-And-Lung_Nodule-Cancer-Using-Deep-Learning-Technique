```python
from keras.models import load_model
classifier = load_model('DCPL.h5')

import numpy as np
from keras.preprocessing import image
import matplotlib.pyplot as plt
import cv2
img_path=input("Enter the absolute path of image")
test_image1= image.load_img(img_path, target_size =(64,64))
test_image= image.load_img(img_path, target_size =(64,64))
#test_image = image.load_img('Dataset/COVID_Test1.png', target_size = (64, 64))
test_image = image.img_to_array(test_image)
test_image = np.expand_dims(test_image, axis = 0)
result = classifier.predict(test_image)

if result[0][2] == 1:

    prediction =  'covid'
    

elif result[0][1] == 1:
 
    prediction =  'pneumonia'
    
elif result[0][0] == 1:
 
    prediction =  'Lung_cancer'

else:
    prediction =  'normal'
    

font = cv2.FONT_HERSHEY_SIMPLEX
texted_image = cv2.putText(img=np.copy(test_image1) , text=prediction, org = (5,10), fontFace=0, fontScale=0.3, color=(225,5,5))
plt.imshow(texted_image)

print(prediction)
```
